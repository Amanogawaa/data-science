```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('./dataset/pokemon.csv')
```

```python
# ?reference
df.head()
```

```python
df['Total'] = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].sum(axis=1)

type_combination = df.groupby(['Type 1', 'Type 2'])['Total'].mean().reset_index()

topPokemon = type_combination.sort_values('Total', ascending=False).head(5)

topPokemon.set_index(['Type 1', 'Type 2']).plot(kind='bar', y='Total' , legend=False)  
```
